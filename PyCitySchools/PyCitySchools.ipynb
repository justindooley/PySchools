```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Finding variables
district_schools = (len(school_data))
district_students = (len(student_data))
district_budget = school_data["budget"].sum()
district_math_mean = student_data["math_score"].mean()
district_reading_mean = student_data["reading_score"].mean()
district_overall_mean = (district_math_mean + district_reading_mean) /2
district_math__passing_amount = (student_data["math_score"] >= 70).sum()
district_math_percent_passing = (district_math__passing_amount / district_students)*100
district_reading__passing_amount = (student_data["reading_score"] >= 70).sum()
district_reading_percent_passing = (district_reading__passing_amount / district_students)*100
district_overall_percent_passing = (district_math_percent_passing + district_reading_percent_passing) /2

# building dataframe
district_summary_df = pd.DataFrame({
    "Total Number of Schools": [district_schools],
    "Total Number of Students": [district_students],
    "Total Budget": [district_budget],
    "Average Math Score": [district_math_mean],
    "Average Reading Score": [district_reading_mean],
    "Overall Average Score": [district_overall_mean],
    "Percentage Passing Math": [district_math_percent_passing],
    "Percentage Passing Reading": [district_reading_percent_passing],
    "Overall Percentage Passing": [district_overall_percent_passing]
})

# cleaner fomatting
district_summary_df["Total Number of Students"] = district_summary_df["Total Number of Students"].map("{:,.0f}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.0f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}".format)
district_summary_df["Overall Average Score"] = district_summary_df["Overall Average Score"].map("{:.2f}".format)
district_summary_df["Percentage Passing Math"] = district_summary_df["Percentage Passing Math"].map("{:.2f}%".format)
district_summary_df["Percentage Passing Reading"] = district_summary_df["Percentage Passing Reading"].map("{:.2f}%".format)
district_summary_df["Overall Percentage Passing"] = district_summary_df["Overall Percentage Passing"].map("{:.2f}%".format)

district_summary_df
```

```python
# Dataframe we are working from
school_data_complete.head()
```

```python
# set index to new df as school name
school_summary = school_data_complete.set_index(["school_name"])
school_summary.head()
```

```python
# create column for school type
school_type = school_data_complete.groupby(["school_name"])["type"].unique()
school_type
```

```python
# create column for total students by school
school_students = school_data_complete.groupby("school_name")['student_name'].count()
school_students = school_students.astype("int")
school_students
```

```python
# create column for Total School Budget
school_budget = school_data_complete.groupby("school_name")['budget'].unique()
school_budget = school_budget.astype('int')
school_budget
```

```python
# create column for Per Student Budget
per_student_budget = school_budget / school_students
per_student_budget
```

```python
# create column for Average Math Score
school_avg_math_score = school_data_complete.groupby(["school_name"])["math_score"].mean()
school_avg_math_score
```

```python
# create column for Average Reading Score
school_avg_reading_score = school_data_complete.groupby(["school_name"])["reading_score"].mean()
school_avg_reading_score
```

```python
# create column for % Passing Math
students_passing_math = school_data_complete[(school_data_complete["math_score"]>=70)]
students_passing_math_per_school = students_passing_math.groupby(["school_name"]).count()["student_name"]
school_percentage_passing_math = (students_passing_math_per_school/school_students)*100
school_percentage_passing_math
```

```python
# create column for % Passing Reading
students_passing_reading = school_data_complete[(school_data_complete["reading_score"]>=70)]
students_passing_reading_per_school = students_passing_reading.groupby(["school_name"]).count()["student_name"]
school_percentage_passing_reading = (students_passing_reading_per_school/school_students)*100
school_percentage_passing_reading
```

```python
# create column for Overall Passing Rate (Average of the above two)
school_overall_passing = (school_percentage_passing_math + school_percentage_passing_reading) / 2
school_overall_passing
```

```python
# Build the dataframes
school_summary = pd.DataFrame({"School Type":school_type,
                                "Total Students": school_students,
                               "Total School Budget": school_budget,
                               "Per Student Budget": per_student_budget,
                               "Average Math Score": school_avg_math_score,
                               "Average Reading Score": school_avg_reading_score,
                               "% Passing Math": school_percentage_passing_math,
                               "% Passing Reading": school_percentage_passing_reading,
                               "% Overall Passing Rate": school_overall_passing
                              })

school_summary2 = pd.DataFrame({"School Type":school_type,
                                "Total Students": school_students,
                               "Total School Budget": school_budget,
                               "Per Student Budget": per_student_budget,
                               "Average Math Score": school_avg_math_score,
                               "Average Reading Score": school_avg_reading_score,
                               "% Passing Math": school_percentage_passing_math,
                               "% Passing Reading": school_percentage_passing_reading,
                               "% Overall Passing Rate": school_overall_passing
                              })

# Format the fancy dataframe
school_summary["Total Students"] = school_summary["Total Students"].map("{:,.0f}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.0f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.0f}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:.2f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:.2f}".format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map("{:.2f}%".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:.2f}%".format)
school_summary["% Overall Passing Rate"] = school_summary["% Overall Passing Rate"].map("{:.2f}%".format)
```
